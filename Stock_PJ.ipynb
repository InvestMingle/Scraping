```python
import pandas as pd
import yfinance as yf

def get_stock_prices(stock_code, start_date, end_date):
    try:
        df = yf.download(stock_code, start=start_date, end=end_date)
        if df.empty:
            raise ValueError(f"No data found for {stock_code} from {start_date} to {end_date}")
    except:
        df = yf.download(stock_code)
    return df['Close']

def get_all_stock_prices(stock_list, start_date, end_date):
    all_prices = pd.DataFrame()

    for index in stock_list:
        stock_info = stock_list[index]
        stock_code = stock_info['code']
        stock_name = stock_info['name']
        try:
            prices = get_stock_prices(stock_code, start_date, end_date)
            all_prices[stock_name] = prices
        except Exception as e:
            print(f"Failed to get data for {stock_name} ({stock_code}): {e}")
            continue

    return all_prices

price_df = get_all_stock_prices(kospi_list, '2013-01-01', '2024-01-01')
price_df.to_csv('kospi_100_10y.csv')
```

```python
import pandas as pd

# Select specific rows and columns from a DataFrame
def select_rows_and_columns(df, index_names):
    df2 = df.set_index(df.columns[0])
    selected_rows = df2.loc[index_names]
    selected_columns = selected_rows[[col for col in selected_rows.columns if '/' in col]]
    return selected_columns

def process_financial_data(stock_code):
    url = f"https://cdn.fnguide.com/SVO2/ASP/SVD_Finance.asp?pGB=1&gicode={stock_code}&cID=&MenuYn=Y&ReportGB=&NewMenuID=103&stkGb=701"
    tables = pd.read_html(url)
    df1_selected = select_rows_and_columns(tables[0], ['매출액', '영업이익', '당기순이익'])
    df2_selected = select_rows_and_columns(tables[2], ['자산', '부채', '자본'])
    merged_df = pd.concat([df1_selected, df2_selected])
    transposed_df = merged_df.T
    return transposed_df

def process_finance_ratio_data(stock_code):
    url = f'https://cdn.fnguide.com/SVO2/ASP/SVD_FinanceRatio.asp?pGB=1&gicode={stock_code}&cID=&MenuYn=Y&ReportGB=D&NewMenuID=104&stkGb=701'
    tables = pd.read_html(url)
    df = select_rows_and_columns(tables[0], ['유동비율계산에 참여한 계정 펼치기', '부채비율계산에 참여한 계정 펼치기',
                                             '영업이익률계산에 참여한 계정 펼치기', 'ROA계산에 참여한 계정 펼치기',
                                             'ROIC계산에 참여한 계정 펼치기'])
    df2 = df.T
    df2.columns = ["유동비율", "부채비율", "영업이익률", "ROA", "ROIC"]
    return df2

def process_invest_index_data(stock_code):
    url = f'https://cdn.fnguide.com/SVO2/ASP/SVD_Invest.asp?pGB=1&gicode={stock_code}&cID=&MenuYn=Y&ReportGB=D&NewMenuID=105&stkGb=701'
    tables = pd.read_html(url)
    df = select_rows_and_columns(tables[3], ['PER계산에 참여한 계정 펼치기','PSR계산에 참여한 계정 펼치기',
                                             'PCR계산에 참여한 계정 펼치기', 'PBR계산에 참여한 계정 펼치기',
                                             '총현금흐름'])
    df2 = df.T
    df2.columns = ['PER', 'PSR', 'PCR', 'PBR', '총현금흐름']
    return df2

# Merge multiple dataframes into one
def merge_dataframes(df_list):
    merged_df = df_list[0]
    for df in df_list[1:]:
        merged_df = merged_df.join(df, how='outer')
    return merged_df

# Get total stock information for a specific stock code
def get_total_stock_inform(code, name):
    df1 = process_financial_data(code)
    df2 = process_finance_ratio_data(code)
    df3 = process_invest_index_data(code)
    merged_df = merge_dataframes([df1, df2, df3])
    merged_df['종목코드'] = code
    merged_df['종목명'] = name
    merged_df = merged_df.reset_index().rename(columns={'index': 'date'})

    # Calculate ROE
    if '당기순이익' in merged_df.columns and '자본' in merged_df.columns:
        merged_df['ROE'] = (merged_df['당기순이익'] / merged_df['자본']) * 100
    else:
        merged_df['ROE'] = None

    return merged_df

# Get stock information for all stock codes in the list
def get_all_stock_inform(code_list):
    total_df = pd.DataFrame()
    for code_info in code_list.values():
        try:
            code = 'A' + code_info['code']
            name = code_info['name']
            df = get_total_stock_inform(code, name)
            total_df = pd.concat([total_df, df], axis=0, ignore_index=True)
        except Exception as e:
            print(f"Failed to get data for {name} ({code}): {e}")
            continue
    return total_df

# Get stock information and save to CSV
stock_inform = get_all_stock_inform(kospi_list)
stock_inform.to_csv('kospi_100_info.csv', index=False)
```
